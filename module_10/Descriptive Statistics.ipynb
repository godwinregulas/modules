#### Finding Mean

```python
import numpy as np
l = [3, 6, 23, 10, 15]
total = 0
for i in l:
    total += i
l_mean = total / len(l)
print(l)
print(l_mean)
print(np.mean(l))
```

#### Finding Median

```python
l = [3, 6, 23, 10, 15]
n = len(l)
sorted_l = sorted(l)
if n % 2 == 0:
    median = (sorted_l[n // 2] + sorted_l[(n // 2) - 1]) / 2
else:
    median = sorted_l[n // 2]
print("Sorted list:", sorted_l)
print("Median:", median)
print(np.median(l)) # Using Numpy
```

#### Finding Mode

```python
l = [3, 6, 23, 10, 15, 6, 10, 6, 10]
freq={}
for key in l:
    if key in freq:
        freq[key] += 1
    else:
        freq[key] = 1
max_count = max(freq.values())
mode = [key for key, value in freq.items() if value == max_count]
print(mode)
```

#### Finding Range

```python
l = [3, 6, 23, 10, 15]
max_val=max(l)
min_val=min(l)
range_val = max_val - min_val
print(range_val)
```

#### Finding Variance

```python
l = [12, 23, 15, 28, 9, 18, 26]
length=len(l)
total=0
for i in l:
    total+=i
l_mean = total/length
sum_squared_diff=0
for i in l:
    sum_squared_diff+= (i - l_mean) ** 2
var = sum_squared_diff/length # Population Variance, For Sample Variance use length-1
print(var)
print(np.var(l))
```

#### Finding Standard Deviation

```python
l=[12, 23, 53, 34, 25, 48, 40, 35]
length=len(l)
total = 0
for i in l:
    total += i
l_mean = total/length
sum_squared_diff=0
for i in l:
    sum_squared_diff += (i - l_mean) ** 2
var=sum_squared_diff/length
std=var ** 0.5
print(std)
print(np.std(l))
```

#### Finding IQR

```python
def find_med(data):
    length = len(data)
    if length % 2 == 0:
        return (data[length//2] + data[(length//2) - 1]) / 2
    else:
        return data[length//2]
l = [12, 23, 53, 34, 25, 48, 40, 35, 45]
sorted_l = sorted(l)
length = len(l)
if length % 2 == 0:
    lower_half = sorted_l[:length//2]
    upper_half = sorted_l[length//2:]
else:
    lower_half = sorted_l[:length//2]
    upper_half = sorted_l[(length//2)+1:]
q1=find_med(lower_half)
q3=find_med(upper_half)
q2=find_med(sorted_l)
iqr=q3-q1
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
print("IQR: ",iqr)
# Using Numpy
l = np.array([12, 23, 53, 34, 25, 48, 40, 35, 45])
# Quartiles
Q1 = np.percentile(l, 25)
Q2 = np.percentile(l, 50)   # Median
Q3 = np.percentile(l, 75)
# IQR
IQR = Q3 - Q1
print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
```
