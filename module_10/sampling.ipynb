# Inferential Statistics
## Sampling Techniques
Imagine you are building a fraud detection ML model.

Population = all transactions in the world

Reality:

95% → Normal transactions

5% → Fraud transactions

Your job:

Pick a training dataset that teaches the model the truth about the world.

```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Created Population
data = pd.DataFrame({
    "transaction_amount" : np.random.normal(100,20,1000),
    "is_fraud" : np.concatenate([
        np.zeros(950), # 95% normal
        np.ones(50) # 5% fraud
    ])
})

data["is_fraud"].value_counts(normalize=True) 
```

#### Stratified Sampling 

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(
    data,
    test_size=0.9, # test data size 90% test and 10% training
    stratify=data["is_fraud"], # ensures the proportion of fraud and not fraud in both datasets test and train
    random_state=42
)

train["is_fraud"].value_counts(normalize=True) # normalizing beacuse we want the proportion and not the count of the data 
# without normalizing we get the counts like 950 and 50 
```

#### Random Sampling

```python
random_sample = data.sample(n=100) # taking 100 sample values from the data
random_sample["is_fraud"].value_counts(normalize=True) # notice the proportion is not the same as the original datasets
```

#### Systematic Sampling

```python
systematic_sample = data.iloc[::10] # taking every tenth data point from the dataseet 
systematic_sample["is_fraud"].value_counts(normalize=True)
```

#### Cluster Sampling

```python
# Create Clusters (hospitals)
hospital_data = pd.DataFrame( {
    "hospital_id" : np.repeat([1,2,3,4,5], 200), # creating 5 clusters each of size 200 
    "age" : np.random.randint(20, 80, 1000), # random ages between 20 and 80 for 1000 patients
    "disease" : np.random.choice([0,1], size=1000, p=[0.95,0.05]) # randomly assign disease status(0 or 1) to each patient
    # 0 will be chosen 95% of time and 1 for the rest 5% times
})

# Select only hospitals 1 and 2 
cluster_sample = hospital_data[hospital_data["hospital_id"].isin([1,2])]  # select all rows in hospitals 1 and 2

cluster_sample["disease"].value_counts(normalize=True) # - This shows the proportion of patients with and without disease in the selected clusters.
```

#### Convinience Sampling

```python
# add a time feature
hospital_data["recent_user"] = np.random.choice([0,1], size=1000, p=[0.7,0.3]) # adds a new coloumn to the dataset assigned 1 for recently visited patients and 0 for rest

# Convinience sample : only recent users
convinience_sample = hospital_data[hospital_data["recent_user"] == 1] # takings rows which satisfies the condition that is : to be a recently visited patient

convinience_sample["disease"].value_counts(normalize=True) # shows the proportion of recently visited patients who have the disease and who doesn't have the disease 
```
